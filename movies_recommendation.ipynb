```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Data Pre-Processing

```python
# loading the data from csv file to the pandas data frame
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
# total number of rows and columns
movies_data.shape
```

```python
# selecting the relevant features for recommendation

selected_features = ['genres', 'keywords', 'original_title', 'tagline', 'director', 'cast']
print(selected_features)
```

```python
# repalcing the null values with null string

for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 selected features

combined_features = movies_data['genres'] +' ' + movies_data['keywords'] + ' '+ movies_data['original_title']+' '+ movies_data['tagline'] + ' '+ movies_data['director'] + ' '+ movies_data['cast']
```

```python
print(combined_features)
```

```python
# converting text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
# getting the similarity scores using similarity

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
# getting the movie name from the user

movie_name = input('Enter your favourite movie name: ')
```

```python
# creating a list with all the movies in the dataset

list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie using title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you: \n ')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]['title'].values[0]
  if (i<31):
    print(i, '.', title_from_index)
    i+=1
```

```python
import requests

API_KEY = "d271f13c8b6be71a5a2bc7ae0e01ab92"
user_mood = "excited"

# Map moods to genres (TMDb genre IDs)
mood_genres = {
    "happy": [35, 12, 10751],   # Comedy, Adventure, Family
    "sad": [18, 10749, 36],     # Drama, Romance, Biography
    "angry": [28, 53, 80],      # Action, Thriller, Crime
    "scared": [27, 53, 9648],   # Horror, Thriller, Mystery
    "relaxed": [99, 16, 14],    # Documentary, Animation, Fantasy
    "excited": [28, 12, 878]    # Action, Adventure, Sci-Fi
}

genre_ids = mood_genres.get(user_mood.lower(), [])

# Discover movies by genre
discover_url = "https://api.themoviedb.org/3/discover/movie"
params = {
    "api_key": API_KEY,
    "with_genres": ",".join(map(str, genre_ids)),
    "sort_by": "popularity.desc",  # sort by popularity
    "page": 1
}

response = requests.get(discover_url, params=params)
movies = response.json().get("results", [])

for movie in movies[:10]:  # show top 10
    print(movie["title"], "-", movie.get("release_date"))
```
